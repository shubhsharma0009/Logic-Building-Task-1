```python
# function to validate recharge
def validate_recharge(amount):

    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least â‚¹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Choose from 199, 299, 399, 599")
        return False

    return True


# main program with retry option
while True:

    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge successful")
        break
    else:
        print("Please try again\n")
```

```python
def check_stock(products):
    for item in products:
        if products[item] < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK")

products = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

check_stock(products)
```

```python
def calculate_fare(km, peak):
    fare = 50 + (12 * km)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    km = float(input("Enter distance in km: "))
    peak = input("Peak hour (yes/no): ")

    total = calculate_fare(km, peak)

    print("Total Fare:", total)

    again = input("Retry? (yes/no): ")

    if again == "no":
        break
```

```python
def check_attendance(att):
    present = 0

    for a in att:
        if a == "P":
            present = present + 1

    percentage = (present / len(att)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P"]

result = check_attendance(attendance)

print("Attendance Percentage:", (attendance.count("P") / len(attendance)) * 100)
print(result)
```

```python
def check_password(password):
    if len(password) < 8:
        print("Weak Password")
        return

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")


pwd = input("Enter password: ")

check_password(pwd)
```

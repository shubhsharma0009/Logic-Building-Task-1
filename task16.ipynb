```python
employees= {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest_score= max(employees.values())

top_performers= []

for name, score in employees.items():
  if score == highest_score:
        top_performers.append(name)

# Step 3: Print result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

```python
sentence="Buy mobile phone buy phone online"
text=sentence.lower()

words=text.split()

frequency={}
for word in words:
  if word in frequency:
      frequency[word] += 1
  else:
      frequency[word]=1
result= {}
for word, count in frequency.items():
  if count > 1:
    result[word]=count
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_count = {}

total = len(emails)

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

total = 0

# calculate total sales
for s in sales:
    total = total + s

# calculate average
average = total / len(sales)

# calculate 30% above average
limit = average + (average * 0.30)

# check spike days
print("Days with sales 30% above average:")
for i in range(len(sales)):
    if sales[i] > limit:
        print(f"Day {i+1}: {sales[i]}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# count each user id
for user in user_ids:
    if user in count:
        count[user] = count[user] + 1
    else:
        count[user] = 1

# print only duplicates
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```

```python
class Light:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(self.name, "is", self.status)

    def turn_off(self):
        self.status = "OFF"
        print(self.name, "is", self.status)



light1 = Light("Bedroom Light")

light1.turn_on()
```

```python
class Employee:

    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept


    def show_id_card(self):

        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)



emp1 = Employee("Rahul", "EMP102", "IT")

emp1.show_id_card()
```

```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone


    def show_contact(self):

        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)



c1 = Contact("Anita", "9876543210")

c1.show_contact()
```

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price


    def show_tag(self):

        print("Product:", self.name)
        print("Price: ₹" + str(self.price))



p1 = Product("Headphones", 2499)

p1.show_tag()
```

```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating


    def show_details(self):

        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")



m1 = Movie("Inception", 4.8)

m1.show_details()
```

```python
class Delivery:

    def __init__(self, name, address):
        self.name = name
        self.address = address


    def show_delivery(self):

        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)



d1 = Delivery("Suman", "Hyderabad")

d1.show_delivery()
```

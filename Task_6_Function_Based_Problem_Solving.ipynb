```python
def parking_system(capacity, logs):

    parked = 0
    peak = 0

    for i in logs:

        if i == "IN":
            parked = parked + 1

        if i == "OUT":
            parked = parked - 1

        if parked > peak:
            peak = parked


    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)



capacity = 50

logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):

    time = distance * 5   # 5 minutes per km

    if traffic == "High":
        time = time + 10
    elif traffic == "Medium":
        time = time + 5
    else:
        time = time + 0


    if weather == "Rainy":
        time = time + 5
    elif weather == "Storm":
        time = time + 10
    else:
        time = time + 0


    print("Estimated Delivery Time:", time, "minutes")



distance = 8
traffic = "High"
weather = "Rainy"

delivery_time(distance, traffic, weather)
```

```python
def theatre_status(total_seats, booked_seats):

    booked = len(booked_seats)

    percent = (booked / total_seats) * 100

    print("Occupancy:", int(percent), "%")

    if percent == 100:
        print("Show Status: Housefull")

    elif percent >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Seats Available")



total_seats = 200

booked_seats = [1] * 150

theatre_status(total_seats, booked_seats)
```

```python
def server_load(cpu):

    total = 0

    for i in cpu:
        total = total + i

    avg = total / len(cpu)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")

    elif avg <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")



cpu = [45, 60, 70, 85, 90]

server_load(cpu)
```

```python
def resource_monitor(data):

    overuse = ""

    for i in data:

        if data[i] > 8:
            overuse = i

    if overuse != "":
        print("Overused Resources:", overuse)
        print("Energy Alert: Yes")

    else:
        print("Overused Resources: None")
        print("Energy Alert: No")



usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
def event_registration(capacity, registrations):

    confirmed = 0
    waitlist = 0

    for i in range(registrations):

        if confirmed < capacity:
            confirmed = confirmed + 1
        else:
            waitlist = waitlist + 1


    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")



capacity = 100
registrations = 105

event_registration(capacity, registrations)
```
